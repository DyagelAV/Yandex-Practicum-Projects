<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> 
    
–•–æ—Ä–æ—à–∏–µ –≥—Ä–∞—Ñ–∏–∫–∏, –º–æ–ª–æ–¥–µ—Ü!

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏–π –≤ –¥–≤—É—Ö –≥—Ä—É–ø–ø–∞—Ö

```python
# –ì–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–µ–∂–¥—É –∫–ª–∞—Å—Å–∞–º–∏ –±—É–¥–µ–º –ø—Ä–æ–≤–µ—Ä—è—Ç—å —Å –ø–æ–º–æ—â—å—é T-test 
# –°–æ–∑–¥–∞–¥–∏–º –ø—Ä–æ—Å—Ç—É—é —Ñ—É–Ω–∫—Ü–∏—é, –∫ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ–º –æ–±—Ä–∞—â–∞—Ç—å—Å—è –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º
# –ü–∞—Ä–∞–º–µ—Ç—Ä equal_var —É–∫–∞–∂–µ–º –∫–∞–∫ False, —É—á–∏—Ç—ã–≤–∞—è, —á—Ç–æ –º—ã –Ω–µ –º–æ–∂–µ–º –æ–¥–Ω–æ–∑–Ω–∞—á–Ω–æ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–≤–Ω—ã, 
# –∞ —Ç–∞–∫–∂–µ –º—ã –∑–Ω–∞–µ–º –æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –≤—ã–±–æ—Ä–æ–∫ (–≤—ã–±–æ—Ä–∫–∏ –Ω–µ —Ä–∞–≤–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π)

def get_ttest_value (data, target):
    results = st.ttest_ind(
    data[data['is_canceled'] == 0][target],
    data[data['is_canceled'] == 1][target],
    equal_var = False)

    alpha = 0.05

    print(results.pvalue)

    if (results.pvalue < alpha):
        print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
    else:
        print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

```python
get_ttest_value(data_train, 'lead_time')
```

**–ü—Ä–æ–≤–µ–¥–µ–Ω –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –ø—Ä–æ—á–∏–º–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ**
- –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω –¥–∏—Å–±–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤, —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ 63% (–Ω–µ –æ—Ç–º–µ–Ω–∏–≤—à–∏–µ –∑–∞–∫–∞–∑) –Ω–∞ 37% (–æ—Ç–º–µ–Ω–∏–≤—à–∏–µ –∑–∞–∫–∞–∑), –¥–∏—Å–∞–±–∞–ª–∞—Å–Ω –Ω—É–∂–Ω–æ —É—á–µ—Å—Ç—å –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –º–æ–¥–µ–ª–∏;
- –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è —Å–≤—è–∑–∏ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –¥—Ä—É–≥–∏–º–∏ –±—ã–ª–∏ —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω—ã –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–æ—Ç–æ—ã—Ä–µ –≤–ª–∏—è—é—Ç –≤ –±–æ–ª—å—à–µ–π —Å—Ç–µ–ø–µ–Ω–∏ (–≤ —Ç.—á. –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∏—Ö —Ç–µ—Å—Ç–æ–≤), —Ç–∞–∫ –∏ –Ω–µ –≤–ª–∏—è—é—â–∏–µ, —Ç–µ–º–Ω–µ-–º–µ–Ω–µ–µ –≤–æ–∑–º–æ–∂–Ω—ã –Ω–µ –ª–∏–Ω–µ–π–Ω–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Å –¥—Ä—É–≥–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏;

–ú–æ–ª–æ–¥–µ—Ü, —á—Ç–æ –ø—Ä–æ–≤–æ–¥–∏—à—å –ø–æ–¥—Ä–æ–±–Ω—ã–π, —Ä–∞–≤–µ—Ä–Ω—É—Ç—ã–π EDA —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –æ—Ç–º–µ–Ω—ã –±—Ä–æ–Ω–∏

### –®–∞–≥ 3. –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ ML-–∑–∞–¥–∞—á–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–∏–∑–Ω–µ—Å-–∑–∞–¥–∞—á–∏

–û—Ü–µ–Ω–∏–º –ø—Ä–∏–±—ã–ª—å –æ—Ç–µ–ª—è –±–µ–∑ –≤–Ω–µ–¥—Ä–µ–Ω–∏—è –¥–µ–ø–æ–∑–∏—Ç–æ–≤. –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é season_ratio, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞—Å—Å—á–∏—Ç–∞–µ—Ç —Å–µ–∑–æ–Ω–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç, 20% –¥–ª—è –æ—Å–µ–Ω–∏, 40% –ª–µ—Ç–æ–º –∏ –¥–ª—è –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–µ–∑–æ–Ω–æ–≤ –æ–Ω —Ä–∞–≤–µ–Ω 1. –ê —Ç–∞–∫–∂–µ —Ñ—É–Ω–∫—Ü–∏—é price_service_room –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –Ω–æ–º–µ—Ä–∞ –∏ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–æ–≤–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è.

```python
def season_ratio(months):

    season_rates = months.copy()
    season_rates.loc[(months == 'June') | (months == 'July') | (months == 'August')] = SUMMER_RATIO
    
    season_rates.loc[(months == 'September') | (months == 'October') | (months == 'November') |
                     (months == 'March') | (months == 'April') | (months == 'May')] = SPRING_RATIO
    
    season_rates.loc[(months == 'December') | (months == 'January') | (months == 'February')] = STANDART_RATIO

    return season_rates
```

–ü—Ä–æ–≤–µ—Ä—å, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –ø—Ä–∏—Å–≤–æ–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –º–µ—Å—è—Ü–∞–º

<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b> –î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –Ω–∞–ø—É—Ç–∞–ª

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ2<a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç</b> 
    

```python
def price_service_room(room_type):

    prices = room_type.copy()
    services = room_type.copy()

    # –ø—Ä–∏—Å–≤–æ–∏–º —Å—Ç–æ–∏–º–æ—Å—Ç—å –Ω–æ–º–µ—Ä–∞
    prices.loc[room_type == 'A'] = PRICE_CLASS_A
    prices.loc[room_type == 'B'] = PRICE_CLASS_B
    prices.loc[room_type == 'C'] = PRICE_CLASS_C
    prices.loc[room_type == 'D'] = PRICE_CLASS_D
    prices.loc[room_type == 'E'] = PRICE_CLASS_E
    prices.loc[room_type == 'F'] = PRICE_CLASS_F
    prices.loc[room_type == 'G'] = PRICE_CLASS_G

    # –ø—Ä–∏—Å–≤–æ–∏–º —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ä–∞–∑–æ–≤–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è
    services.loc[room_type == 'A'] = SERVICE_CLASS_A
    services.loc[room_type == 'B'] = SERVICE_CLASS_B
    services.loc[room_type == 'C'] = SERVICE_CLASS_C
    services.loc[room_type == 'D'] = SERVICE_CLASS_D
    services.loc[room_type == 'E'] = SERVICE_CLASS_E
    services.loc[room_type == 'F'] = SERVICE_CLASS_F
    services.loc[room_type == 'G'] = SERVICE_CLASS_G

    return prices, services
```

–û–ø—Ä–µ–¥–µ–ª–µ–∏–º —Å–µ–∑–æ–Ω–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ –≤—ã–±–æ—Ä–∫–∏, —Å—Ç–æ–∏–º–æ—Å—Ç—å –Ω–æ–º–µ—Ä–∞ –∏ —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ä–∞–∑–æ–≤–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è. 

```python
season_rates = season_ratio(data_test['arrival_date_month'])
room_price, services = price_service_room(data_test['reserved_room_type'])
total_nights = data_test['total_nights']
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º –ø—Ä–∏–±—ã–ª—å –æ—Ç–µ–ª—è –∏–ª–∏ —É–±—ã—Ç–æ–∫, –≤ —Å–ª—É—á–∞–µ –æ—Ç–º–µ–Ω—ã –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è. –¢–∞–∫ –∫–∞–∫ —É–±–æ—Ä–∫–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è –∫–∞–∂–¥—ã–µ –¥–≤–∞ –¥–Ω—è, —Ä–∞–∑–¥–µ–ª–∏–º –ø–æ–ø–æ–ª–∞–º –∏ –æ–∫—Ä—É–≥–ª–∏–º –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ—á–µ–π.

```python
profit = data_test['arrival_date_month'].copy()

profit.loc[data_test['is_canceled'] == 0] = \
    room_price * season_rates * total_nights - services * np.ceil(total_nights / 2)

profit.loc[data_test['is_canceled'] == 1] = - (room_price * season_rates + services)
```

–í—ã—á–∏—Å–ª–∏–º –ø—Ä–∏–±—ã–ª—å –∏ —É–±—ã—Ç–∫–∏ –æ—Ç–µ–ª—è –±–µ–∑ —É—á—ë—Ç–∞ –≤–Ω–µ–¥—Ä–µ–Ω–∏—è –¥–µ–ø–æ–∑–∏—Ç–æ–≤.

```python
profit_total = profit.sum()
loss_total = profit.loc[data_test['is_canceled'] == 1].sum()

print(f"–ü—Ä–∏–±—ã–ª—å –æ—Ç–µ–ª–µ—è –±–µ–∑ –≤–Ω–µ–¥—Ä–µ–Ω–∏—è –¥–µ–ø–æ–∑–∏—Ç–æ–≤: {profit_total:,} —Ä—É–±.")
print(f"–£–±—ã—Ç–∫–∏ –æ—Ç–µ–ª—è –ø—Ä–∏ –æ—Ç–º–µ–Ω–µ –±—Ä–æ–Ω–∏: {loss_total:,} —Ä—É–±.")
print(f"–£–±—ã—Ç–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç {-1 * loss_total / profit_total:.1%} –æ—Ç –ø—Ä–∏–±—ã–ª–∏.")
```

- –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω—ã —É–±—ã—Ç–∫–∏ –æ—Ç–µ–ª—è –ø—Ä–∏ –æ—Ç–º–µ–Ω–µ –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤–µ—Å—å–º–∞ –Ω–µ —É—Ç–µ—à–∏—Ç–µ–ª—å–Ω—ã–µ, —É–±—ã—Ç–∫–∏ —Å–æ—Å—Ç–∞–≤–ª–∏ –±–æ–ª–µ–µ 50% –æ—Ç –ø—Ä–∏–±—ã–ª–∏.

```python
num_columns_list = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'previous_cancellations',
                    'previous_bookings_not_canceled', 'booking_changes', 'days_in_waiting_list',
                    'total_of_special_requests', 'total_nights']

cat_columns_list = ['arrival_date_month', 'arrival_date_week_number',
                    'arrival_date_day_of_month', 'adults', 'children', 'babies', 'meal',
                    'country', 'distribution_channel', 'is_repeated_guest', 'reserved_room_type',
                    'customer_type',  'required_car_parking_spaces']
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –∏ –Ω–∞–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –º–æ–¥–µ–ª–µ–π.

```python
# —Å–æ–∑–¥–∞–¥–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–µ–π
features_train = data_train.drop('is_canceled', axis=1)
features_test = data_test.drop('is_canceled', axis=1)

# –≤—ã–¥–µ–ª–∏–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
target_train = data_train['is_canceled']
target_test = data_test['is_canceled']

# –∑–∞–¥–∞–¥–∏–º –¥–∏–∞–ø–∞–∑–æ–Ω—ã –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
#param_distributions_rfc = {
#                       'RFC__n_estimators': range(1, 100, 1),
#                       'RFC__max_depth': range(1, 50, 1),
#                       'RFC__min_samples_split': range(2, 52, 1),
#                       'RFC__min_samples_leaf': range(2, 52, 1),
#                       'RFC__criterion':['gini','entropy']}


param_distributions_rfc = {'RFC__n_estimators': range(50, 325, 25),
                       'RFC__max_depth': range(1, 18, 1),
                       'RFC__min_samples_split': range(2, 52, 5),
                       'RFC__min_samples_leaf': range(2, 52, 5)}
param_distributions_lr = {
                       'LR__penalty':['l1','l2'],
                       'LR__C': np.arange(0.1,5, 0.1)}

# –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –Ω–∞ 5 –≤—ã–±–æ—Ä–∫–∞—Ö
kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=STATE)
```

–î–ª—è –æ–±—É—á–µ–Ω–∏—è –≤—ã–±–µ—Ä–µ–º –º–µ—Ç—Ä–∏–∫—É recall (–ø–æ–ª–Ω–æ—Ç–∞), –ø–æ—Ç–æ–º—É —á—Ç–æ —ç—Ç–∞ –º–µ—Ç—Ä–∏–∫–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é –æ–±—ä–µ–∫—Ç–æ–≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –∏–∑ –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –Ω–∞—à–µ–ª –∞–ª–≥–æ—Ä–∏—Ç–º. –î–ª—è —Å–Ω–∏–∂–µ–Ω–∏—è –∑–∞—Ç—Ä–∞—Ç –Ω–∞–º –∫–∞–∫ —Ä–∞–∑ –≤–∞–∂–Ω–æ –Ω–∞–π—Ç–∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ—Ö, –∫—Ç–æ –æ—Ç–º–µ–Ω–∏—Ç –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –ø—Ä–µ–¥–ª–æ–∂–∏—Ç—å –∏–º –≤–Ω–µ—Å—Ç–∏ –¥–µ–ø–æ–∑–∏—Ç.

–û–±—É—á–∏–º –º–æ–¥–µ–ª–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è, –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –±–∞–ª–∞–Ω–∞—Å–∏—Ä–æ–≤–∫–∏ –∫–ª–∞—Å—Å–æ–≤.

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> 
    
    
–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –µ—Å—Ç—å –æ–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∞ –º–µ—Ç—Ä–∏–∫–∏.

#### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å —Å –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–æ–π –≤–µ—Å–æ–≤ c –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º —Ç–µ—Ö–Ω–∏–∫–∏ OHE –∏ StandardScaler –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö

```python
FPP = ('FPP', ColumnTransformer([
        ('num_features', make_pipeline(StandardScaler()), num_columns_list),
        ('cat_features', make_pipeline(OneHotEncoder(sparse=False, handle_unknown='ignore')), cat_columns_list)]))
RFC = ('RFC', RandomForestClassifier(random_state=STATE, class_weight='balanced'))

# –∑–∞–¥–∞–¥–∏–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–±—Ä–∞–±–æ—Ç–∫–∏
pipeline_RFC = Pipeline(steps=[FPP, RFC])

rs_rfc_ohe_ss = RandomizedSearchCV(
    pipeline_RFC, 
    param_distributions_rfc, 
    n_iter=100, 
    cv=kf, 
    verbose=1,
    n_jobs=-1, 
    scoring='recall', 
    return_train_score=True, 
    random_state=STATE)

start_time = time.time()
rs_rfc_ohe_ss.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

print(f'recall = {rs_rfc_ohe_ss.best_score_:.4f}')
df_results = pd.DataFrame({'–ú–æ–¥–µ–ª—å': ['Random Forest + OHE + Standart Scaler'], 
                           'recall': [rs_rfc_ohe_ss.best_score_],
                           '–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è (–º–∏–Ω—É—Ç—ã)': [run_time], 
                           '–ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã': [rs_rfc_ohe_ss.best_params_]})
```

–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –∫–∞—á–µ—Å—Ç–≤–æ –∏—Å—Å–ª–µ–¥—É–µ—à—å –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏, –∞ —Ç–∞–∫–∂–µ –ø–æ–¥–±–∏—Ä–∞–µ—à—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

–£ OneHotEncoder –º–æ–∂–Ω–æ –∑–∞–¥–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä *drop=first*, —á—Ç–æ–±—ã —É—á–µ—Å—Ç—å –ø—Ä–æ–±–ª–µ–º—É –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç–∏.
        
–ß—Ç–æ–±—ã —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞–±–æ—Ç–∞–ª –≤–º–µ—Å—Ç–µ —Å *handle_unknown* –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ–±–Ω–æ–≤–∏—Ç—å –≤–µ—Ä—Å–∏—é sklearn

–í–∞–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —ç—Ç–æ –¥–æ –∏–º–ø–æ—Ä—Ç–∞ –∫–æ–¥–µ—Ä–∞.

#### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å —Å –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–æ–π –≤–µ—Å–æ–≤ c –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º —Ç–µ—Ö–Ω–∏–∫–∏ OrdinalEncoder –∏ MinMaxScaler –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö

```python
FPP = ('FPP', ColumnTransformer([
        ('num_features', make_pipeline(MinMaxScaler()), num_columns_list),
        ('cat_features', make_pipeline(OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)), cat_columns_list)]))

RFC = ('RFC', RandomForestClassifier(random_state=STATE, class_weight='balanced'))

# –∑–∞–¥–∞–¥–∏–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–±—Ä–∞–±–æ—Ç–∫–∏
pipeline_RFC = Pipeline(steps=[FPP, RFC])

rs_rfc_mms_oe = RandomizedSearchCV(
    pipeline_RFC, 
    param_distributions_rfc, 
    n_iter=100, 
    cv=kf, 
    verbose=1,
    n_jobs=-1, 
    scoring='recall', 
    return_train_score=True, 
    random_state=STATE)

start_time = time.time()
rs_rfc_mms_oe.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

print(f'recall = {rs_rfc_mms_oe.best_score_:.4f}')

df_results.loc[len(df_results.index)] = ['Random Forest + OE + Min Max Scaler',
                                         rs_rfc_mms_oe.best_score_, run_time, rs_rfc_mms_oe.best_params_]
```

```python
FPP = ('FPP', ColumnTransformer([
        ('num_features', make_pipeline(StandardScaler()), num_columns_list),
        ('cat_features', make_pipeline(OneHotEncoder(sparse=False, handle_unknown='ignore')), cat_columns_list)]))

LR = ('LR', LogisticRegression(random_state=STATE, class_weight='balanced'))

# –∑–∞–¥–∞–¥–∏–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–±—Ä–∞–±–æ—Ç–∫–∏
pipeline_LR = Pipeline(steps=[FPP, LR])

rs_lr_ss_ohe = RandomizedSearchCV(
    pipeline_LR, 
    param_distributions_lr, 
    n_iter=100, 
    cv=kf, 
    verbose=1,
    n_jobs=-1, 
    scoring='recall', 
    return_train_score=True, 
    random_state=STATE)

start_time = time.time()
rs_lr_ss_ohe.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

print(f'recall = {rs_lr_ss_ohe.best_score_:.4f}')

df_results.loc[len(df_results.index)] = ['Logistic Regression + OHE + Standart Scaler',
                                         rs_lr_ss_ohe.best_score_, run_time, rs_lr_ss_ohe.best_params_]
```

–ü–æ—Å—Ç—Ä–æ–∏–º ROC-–∫—Ä–∏–≤—É—é.

```python
# –≤—ã–≤–µ–¥–µ–º ROC-–∫—Ä–∏–≤—É—é
fpr, tpr, thresholds = roc_curve(target_test, rf_probs)
plt.figure(figsize=(15, 10))
plt.plot(fpr, tpr, label='MinMaxScaler + OrdinalEncoder + Random Forest')

# ROC-–∫—Ä–∏–≤–∞—è —Å–ª—É—á–∞–π–Ω–æ–π –º–æ–¥–µ–ª–∏ (–≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –ø—Ä—è–º–∞—è)
plt.plot([0, 1], [0, 1], linestyle='--', label='RandomModel')

# —É—Å—Ç–∞–Ω–æ–≤–∏–º –≥—Ä–∞–Ω–∏—Ü—ã –æ—Å–µ–π –æ—Ç 0 –¥–æ 1
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

plt.xlabel('False Positive Rate', color='grey', weight = 'bold')
plt.ylabel('True Positive Rate', color='grey', weight = 'bold')
plt.title('ROC-–∫—Ä–∏–≤–∞—è', weight = 'bold')
plt.legend()
plt.show()
```

–ü—Ä–æ–≤–µ—Ä–∏–º –º–æ–¥–µ–ª—å –Ω–∞ –∞–¥–µ–∫–≤–∞—Ç–Ω–æ—Å—Ç—å. –°–æ–∑–¥–∞–¥–∏–º baseline –º–æ–¥–µ–ª—å —Å –ø–æ–º–æ—â—å—é Dummy-–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞. –†–∞—Å—Å—á–∏—Ç–∞–µ–º –¥–ª—è —ç—Ç–æ–π –º–æ–¥–µ–ª–∏ –º–µ—Ç—Ä–∏–∫—É recall –∏ —Å—Ä–∞–≤–Ω–∏–º –µ—ë —Å –Ω–∞—à–µ–π –º–æ–¥–µ–ª—å—é.

```python
dc = DummyClassifier(strategy='stratified', random_state=STATE)
dc.fit(features_train, target_train)
dc_pred = dc.predict(features_test)
dc_recall = recall_score(target_test, dc_pred)
# –≤—ã–≤–µ–¥–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è
print(f'–ü–æ–ª–Ω–æ—Ç–∞ –¥–ª—è –º–æ–¥–µ–ª–∏ Dummy-–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {dc_recall:.4f}')
```

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–π –º—ã –ø–æ–ª—É—á–∏–ª–∏ –º–æ–¥–µ–ª—å —Å —Ö–æ—Ä–æ—à–∏–º–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º–∏ –º–µ—Ç—Ä–∏–∫. –ú–æ–¥–µ–ª—å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –ª—É—á—à–µ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç, —á–µ–º –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è. –î–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏ 0.6001, —á—Ç–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –≤—ã—à–µ, —á–µ–º —É baseline –º–æ–¥–µ–ª–∏ 0.3707.
