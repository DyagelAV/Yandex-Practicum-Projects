Основные функции

```python
# Функция печати гистограммы и диаграммы размаха
def get_hist_and_boxplot (data, target, bins):  
    sns.set_style("white")
    fig, ax = plt.subplots(figsize=(18,5))
    sns.histplot(data=data, x=target, kde=True, bins=bins, edgecolor='black', ax=ax)
    plt.title('Гистограмма', fontsize='14')
    plt.xlabel(target, fontsize='12')
    plt.ylabel('Частота (количество)', fontsize='12')
    plt.show()
    
    fig, ax = plt.subplots(figsize=(18,2))
    sns.boxplot(data=data,x=target, ax=ax)
    #plt.title('Диаграмма размаха', fontsize='14')
    #plt.xlabel(target, fontsize='12')
    plt.show()

def get_dict_and_fillna(data, column_to_index, target, agg):
    # .....
    pivot = pd.pivot_table(data, index = column_to_index, values = target, aggfunc = agg)
    #Сводную таблицу переведем в словарь
    dict_state = dict(pd.Series(pivot[target].values, index=pivot.index))
    # .....
    data.loc[data[target].isna(), target] = data[column_to_index].map(dict_state)
    return data

def get_cat_body(data):
    if 'cab' in str(data) or 'supercrew' in str(data): return 'pickup' # supercrew это picup от Ford
    elif 'convertible' in str(data): return 'convertible'
    elif 'coupe' in str(data) or 'koup' in str(data): return 'coupe'
    elif 'wagon' in str(data): return 'wagon'
    elif 'sedan' in str(data): return 'sedan'
    elif 'van' in str(data): return 'van'
    elif 'suv' in str(data): return 'suv'
    elif 'hatchback' in str(data): return 'hatchback'
    else: return data

def get_cat_odometer(data):
    if data <= 1000: return 'до 1 тыс.'
    elif data <= 15000: return 'до 15 тыс.'
    elif data <= 30000: return 'до 30 тыс.'
    elif data <= 45000: return 'до 45 тыс.'
    elif data <= 60000: return 'до 60 тыс.'
    elif data <= 75000: return 'до 75 тыс.'
    elif data <= 90000: return 'до 90 тыс.'
    elif data <= 120000: return 'до 120 тыс.'
    elif data <= 150000: return 'до 150 тыс.'
    elif data <= 200000: return 'до 200 тыс.'
    else: return 'свыше 200 тыс.'


def get_color(data):
    if 'gold' in str(data): return 'other' 
    elif 'brown' in str(data): return 'other' 
    elif 'green' in str(data): return 'other'
    elif 'beige' in str(data): return 'other' 
    elif 'orange' in str(data): return 'other' 
    elif 'off-white' in str(data): return 'other' 
    elif 'burgundy' in str(data): return 'other' 
    elif 'yellow' in str(data): return 'other' 
    elif 'charcoal' in str(data): return 'other' 
    elif 'purple' in str(data): return 'other' 
    elif 'turquoise' in str(data): return 'other' 
    elif 'lime' in str(data): return 'other' 
    elif 'pink'in str(data): return 'other' 
    else: return data

def get_color_interior(data):
    if 'black' in str(data): return data
    elif 'beige' in str(data): return data
    elif 'gray' in str(data): return data
    elif 'tan' in str(data): return data
    elif 'brown' in str(data): return data
    elif 'unknown' in str(data): return data
    else: return 'other'

def get_is_summer(data):
    if data == 5 or data == 6 or data == 7 or data == 8: return True
    else: return False
```

### Предварительная обработка данных

**Стртатегия работы с пропусками**

**Дополнительные признаки**

**Признак Condition** - состояние автомобилей на момент аукциона. Как ранее установили - доля пропусков составялет 2.1%. Брать среднее/ медиану по моделе / марке авто не совсем правильно - Форд фокус может быть как в отличном состоянии, так и "убитый". При этом в данных есть признак state - состояние, в котором автомобиль выставлен на аукцион. Сделаем допущение, что состояние при кором автомобиль выставлен и состояние на момент аукциона +/- не изменно и заполнять пропуски по столбцу condition будем медианным значением столбца state.

**Признак Odometer** - заполнять данный признак будем в логике, котоаря приведена выше для признака 

**Признак transmission** - заполонение пропусков по модели авто и  уровню отделки салона не получится, так как одна и таже модель авто при одинаковых значениях trim (уровень отделки салона) может иметь разные типы трансмиссии. Предлагется заполнять пропуски значением "unknown"  

**Признак color и interior** - в процессе анализа неявных дубликатов не выявлен, однако указан знак "-" - вероятнее всего означает не указанное значение. Восстановить  цвета салона и интерьера по имеющимся данным весьма затруднительно (если вообще возможно), а учитывая, что доля пропусков составялет менее 1% заменим прпоуски ,а также знак "-" на "unknown.   

**Признак Make** - в данных присутствуют неявные дубликаты (скоращенные и полные названия марок авто, не полные названия и др.) Для чистоты в данных следуюет устранить этот момент. Также в данных есть марки авто с препиской Truck или Tk, что могло указать на факт большого типа машин, как правило, truck это "пикап". Сущестовавла гипотеза, если в столбце Make указано Truck или Tk, то это пикап, оданко такая гипотеза не подтвредилась. В данных есть авто без указания Truck или Tk и они являются пикапами (проверено вручную), тем более в обучающем наборе данных по столбцу Make есть данные по Mazda Tk, при этом модель СХ-7, которая явялется кроссовером, а не пикапом. У Mazda есть truck (пикап), но модель совершенно другая. Поэтому выделять отдельный признак "is_truck" не стали. Для того, чтобы избавитсья от неявных дубликатов приведем все к верхнему регистру, затем выполним ряд преобразований.

```python
def data_preprocessing(data):
    
    # ===== первым шагом приведем дату к удобному виду =====
    data['saledate'] = pd.to_datetime(data['saledate'].apply(lambda x: x.split(' G', 1)[0]), format = '%a %b %d %Y %H:%M:%S').dt.round('1d')

    # Устранение неявных дубликатов
    list_to_preprocessing = ['make', 'model', 'trim', 'transmission', 'color', 'interior', 'seller', 'body'] 
    for i in list_to_preprocessing: data[i] = data[i].str.lower().str.strip()
    
    # Заполнение пропусков
    for i in ['condition', 'odometer']: data = get_dict_and_fillna(data, 'year', i, 'median')
    for i in list_to_preprocessing: data[i] = data[i].fillna('unknown') # попытка наиболее часто встрчющемся призанком заполнить ??? смотри ссылку ниже
    # https://www.dmitrymakarov.ru/data-analysis/nan-06/#20-zapolnenie-naibolee-chastotnym-znacheniem
    for i in ['color', 'interior']: data[i] = data[i].replace(['—'], ['unknown'])
            
    #Выполянем замену для двойных названий, сокращенных
    data['make'] = data['make'].replace(['mercedes', 'mercedes-b', 'vw', 'ram'],['mercedes-benz','mercedes-benz','volkswagen','dodge'])
    #Выполнем преобразование для авто у которых в названии присутствует Tk или Truck
    data['make'] = data['make'].replace(list(data['make'].unique()), [str(x).split(' ', 1)[0] for x in data['make'].unique()])
    
    # Введение доп. признаков
    data['body_cat'] = data['body'].apply(get_cat_body)
    data['sale_year'] = data['saledate'].dt.year
    data['sale_month'] = data['saledate'].dt.month
    data['sale_day'] = data['saledate'].dt.day
    
    return data
```

```python
data_train = data_preprocessing(data_train)
```

Изучив срезы по автомобилям с пробегом свыше 600 тыс км. можно сделать наблюдение, что пробег 999 999 км указан как у автомобилей выпущенных начиная с 2000 г., так и до 2000г., вероятнее всего 999 999 означет, что продавец не указал пробег и система автоматически проставила такие значения (это можно уточнить у заказчика) или другая прчина, естественно эти данные удаляться не будут, т.к. и на тестовой выборке можно встретить такие значения. Заменять их на нулевые не самый правильный подход. Проверим гипотез о равенстве средней цены продажи для авто с пробегом свыше 450 тыс и автомобилей с пробегом 999 тыс. Значение 450 выбрано исхояд из того, что согласно диаграмме размаха примерно с этого значения начинаются крайне редкие значения.

```python
results = st.ttest_ind(
    data_train.query("odometer > 450000 and odometer < 999999")['sellingprice'],
    data_train.query("odometer == 999999")['sellingprice'],
    equal_var = False)

alpha = 0.05

print(results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Учитывая, что нулевую гиоптезу отвергнуть не удлаось, а значит средняя цена для таких равна, то в целом предлагается ограничить значения пределом в 450 тыс км. Посмотрим количество записей при пробеге равном 1 км.
